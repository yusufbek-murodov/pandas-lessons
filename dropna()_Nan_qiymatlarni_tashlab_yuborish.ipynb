```python
import pandas as pd
```

Yangi DF yaratamiz.Jadval tarkibidagi **NaN** qiymatlariga ahamiyat bering.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

## **.isnull()**

**.isnull()** metodi yordamida ustun qiymati **NaN** ga teng ( True ) yoki ( False ) ekanini tekshiramiz. Ya'ni, NaN qiymat o'rnida **True** ko'ramiz.

```python
df.isnull()
```

```python
## Boolean qiymati True=1 va False=0 bo'lgani uchun .sum() metodi yordamida har ustundagi jami NaN qiymatlar sonini ko'rishimiz mumkin.
```

```python
print(df.isnull().sum())
```

```python
nulls = df.isnull().sum() # NaN qiymatlar soni
rows = df.shape[0] # qatorlar soni
(nulls/rows)*100 # NaN qiymatlarning foizdagi qiymati
```

Yuqoridagi ba'zi ustunlardagi **NaN** qiymatlar soni 25% dan ham ko'p ekanini ko'rishimiz mumkin. Bu, albatta yaxshi emas.

**pandas** da **NaN** qiymatlar bilan ishlash uchun bir nechta metodlar bor:

- **dropna** - **NaN** qiymatlar bor ustun yoki qatorlarni o'chirib tashlash

- **fillna** - **NaN** qiymatlarni boshqa qiymatlar yordamida to'ldirib ketish (misol uchun avvalgi yoki keyingi qator qiymatlari)

- **isnull** - **NaN** qiymatlarni filtrlash (NaN qiymat True qaytaradi)

- **notnull** - **isnull** ning aksi


**.dropna() NaN qiymatlarni tashlab yuborish**

Ushbu metod yordamida **NaN** qiymatlar bor ustun yoki qatorlarni butunlay tashlab yuborish mumkin.
