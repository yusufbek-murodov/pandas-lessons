```python
import pandas as pd
import numpy as np
```

```python
# Hierarchical indexing

data = pd.Series(np.random.randn(9),
                 index=['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd']) # odatiy indeksli Ser
data.index
```

```python
data
```

```python
data1 = pd.Series(np.random.randn(9),
                 index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],[1, 2, 3, 1, 3, 1, 2, 2, 3]]) # irarxik indeksli Ser
data1.index
```

```python
data1['c': 'd']
```

```python
data1.loc[:, 2] # barcha 2-indeksdagi ma'lumotlarni ajratib beradi
```

```python
data1.loc[['b', 'd']] # barcha b va d indeksdagi elementlar
```

```python
# Unstack
```

```python
data1
```

```python
data1.unstack() # sochilgan Ser.ni df ko'rinishida qaytaradi
```

```python
data1.unstack().stack() # df.ni yig'ilgan Ser.ga o'giradi
```

```python
df = pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
df
```

```python
df.info()
```

```python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']

df.columns = columns
```

```python
df.head()
```

```python
df['Region'] = df['Region'].map(lambda x: x.replace('"', '')) # Region ustuniga tegishli ma'lumotlar
df
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"', ''))
df
```

```python
df['Country'] = df['Country'].map(lambda x: x.strip())
df
```

```python
df.to_csv("Updated_dataset.csv", index=False)
```
